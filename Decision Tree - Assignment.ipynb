```python
#Question 1: What is a Decision Tree, and how does it work in the context of classification?
'''Answer:  A Decision Tree is a non-linear machine learning model that represents decisions and their possible consequences in a tree-like structure. In classification, it works by recursively partitioning the dataset into smaller, more homogeneous subsets—often called "pockets"—based on nested if-else conditions. 
The process starts at a root node (the first decision point) and moves through internal nodes (where further decisions are made) until it reaches a leaf node, which provides the final class label. 
For instance, to classify if a person is a "fraudster," the tree might first ask if their income is above $30k; if yes, it might then check if their age is over 35. The goal is to continue splitting until the leaf nodes are as "pure" as possible, meaning every data point in that node belongs to the same class.
'''
```

```python
#Question 2: Explain the concepts of Gini Impurity and Entropy as impurity measures. How do they impact the splits in a Decision Tree? 
'''Answer : Gini Impurity and Entropy are mathematical formulas used to measure the "disorder" or impurity within a node. 
Gini Impurity calculates the probability of a random element being misclassified; it ranges from 0 (pure) to 0.5 (equally mixed). 
Entropy measures the randomness of the data; a pure node has an entropy of 0, while a totally random 50/50 split has an entropy of 1. 
These measures impact splits because the algorithm evaluates every possible feature and split point, choosing the one that results in the lowest combined impurity for the child nodes. In simple terms, the tree "prefers" the split that makes the resulting groups as clean and organized as possible. 
Gini is often used by default because it is computationally faster since it does not require logarithmic calculations.
'''
```

```python
#Question 3: What is the difference between Pre-Pruning and Post-Pruning in Decision Trees? Give one practical advantage of using each. 
'''Answer :  Pruning is the process of trimming a tree to prevent it from "overfitting"—a situation where the model memorizes the training data too well but fails on new data. 
Pre-pruning stops the tree from growing before it reaches its full size by setting "stopping criteria," such as a maximum tree depth or a minimum number of samples required to split. A practical advantage is that it saves time and memory during training. 
Post-pruning allows the tree to grow to its maximum length (where every leaf is pure) and then cuts back branches that add little predictive power. 
A practical advantage is that it can capture complex patterns first and then simplify, which is often more accurate for smaller datasets than stopping early.
'''
```

```python
#Question 4: What is Information Gain in Decision Trees, and why is it important for choosing the best split?
'''Answer :  Information Gain (IG) is the statistical measure of the reduction in entropy (uncertainty) achieved by splitting the data on a specific attribute. 
It is calculated by taking the entropy of the parent node and subtracting the weighted average entropy of the child nodes. This is important because it serves as the selection guide during tree construction: the algorithm calculates the IG for every available feature and value, then selects the one with the highest gain to perform the split.
For example, if "Gender" reduces uncertainty about a outcome more than "Class" does, the tree will split on Gender first. Essentially, it ensures the tree stays on the "optimal path" by asking the most useful #Questions first.
'''
```

```python
#Question 5: What are some common real-world applications of Decision Trees, and what are their main advantages and limitations? 
'''Answer : Decision Trees are used in Finance for credit risk assessment and fraud detection, in Healthcare for diagnosing conditions based on symptoms, and in Marketing for customer segmentation. 
Their main advantages are that they are "simple to understand" (they mirror human logic), require minimal data cleaning, and can capture non-linear relationships. 
However, their main limitations include being prone to overfitting and being unstable (small changes in data can create a completely different tree). They are also "greedy algorithms," meaning they make the best choice at each step without considering if a different choice might be better for the overall tree structure.
'''
```

```python
'''Dataset Info: 
   -Iris Dataset for classification tasks (sklearn.datasets.load_iris() or provided CSV).
   -Boston Housing Dataset for regression tasks (sklearn.datasets.load_boston() or provided CSV).
'''
#Question 6: Write a Python program to: 
'''-Load the Iris Dataset 
   -Train a Decision Tree Classifier using the Gini criterion 
   -Print the model’s accuracy and feature importances'''
'''Answer: 
'''
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


iris = load_iris()  # Load data and fit model
X, y = iris.data, iris.target
model = DecisionTreeClassifier(criterion='gini').fit(X, y)

print(f"Accuracy: {accuracy_score(y, model.predict(X))}")
print(f"Feature Importances: {model.feature_importances_}")
'''Note: This code assumes training on the full set as per typical assignment requests.'''
```

```python
#Question 7: Write a Python program to:
'''> Load the Iris Dataset 
   > Train a Decision Tree Classifier with max_depth=3 and compare its accuracy to a fully-grown tree.'''
'''Answer :
Python'''

tree_lim = DecisionTreeClassifier(max_depth=3).fit(X, y)  # Limited depth vs Fully-grown
tree_full = DecisionTreeClassifier().fit(X, y)

print(f"Depth 3 Accuracy: {accuracy_score(y, tree_lim.predict(X))}")
print(f"Full Tree Accuracy: {accuracy_score(y, tree_full.predict(X))}")
```

```python
#Question 8: Write a Python program to: 
'''-Load the Boston Housing Dataset 
   -Train a Decision Tree Regressor 
   -Print the Mean Squared Error (MSE) and feature importances'''
'''Answer :
Code'''
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.datasets import fetch_california_housing # Boston is deprecated

data = fetch_california_housing()
model = DecisionTreeRegressor().fit(data.data, data.target)
mse = mean_squared_error(data.target, model.predict(data.data))

print(f"MSE: {mse}")
print(f"Feature Importances: {model.feature_importances_}")
```

```python
#Question 9: Write a Python program to: 
'''-Load the Iris Dataset 
   -Tune the Decision Tree’s max_depth and min_samples_split using GridSearchCV 
   -Print the best parameters and the resulting model accuracy'''
'''Answer : 
Code'''
from sklearn.model_selection import GridSearchCV

params = {'max_depth': [2, 3, 4, 5], 'min_samples_split': [2, 5, 10]}
grid = GridSearchCV(DecisionTreeClassifier(), params).fit(X, y)

print(f"Best Params: {grid.best_params_}")
print(f"Best Score: {grid.best_score_}")
```
